1.  What is hypothesis testing in statistics?
  - Hypothesis testing is a statistical method used to make decisions or draw conclusions about a population based on sample data. It involves formulating two competing hypotheses:

Null Hypothesis (H0): This is the initial assumption or the statement that we want to test. It usually represents the status quo or no effect.
Alternative Hypothesis (H1 or Ha): This is the opposite of the null hypothesis and represents the claim we are trying to find evidence for.
Steps in Hypothesis Testing

Formulate the Hypotheses: Clearly state the null and alternative hypotheses.
Set the Significance Level: Determine the alpha level (α), which represents the probability of rejecting the null hypothesis when it is actually true. Commonly used values for alpha are 0.05 or 0.01.
Collect and Analyze Data: Gather data from a sample and calculate the relevant test statistic.
Determine the p-value: The p-value is the probability of obtaining the observed data (or more extreme data) if the null hypothesis is true.
Make a Decision: Compare the p-value to the significance level. If the p-value is less than or equal to alpha, reject the null hypothesis in favor of the alternative hypothesis. Otherwise, fail to reject the null hypothesis.
Example

Suppose we want to test if a new drug is effective in lowering blood pressure.

Null Hypothesis (H0): The drug has no effect on blood pressure.
Alternative Hypothesis (H1): The drug lowers blood pressure.

3. What is the significance level in hypothesis testing, and why is it important?
  What is the Significance Level?

The significance level, often denoted by alpha (α), is the probability of rejecting the null hypothesis when it is actually true. In simpler terms, it represents the risk of concluding that an effect exists when it doesn't.

Why is it Important?

The significance level is crucial because it helps us control the risk of making a Type I error. A Type I error occurs when we reject the null hypothesis when it is actually true.

Controlling Type I Error: By setting a significance level, we establish a threshold for rejecting the null hypothesis. If the p-value (the probability of observing the data or more extreme data if the null hypothesis is true) is less than or equal to the significance level, we reject the null hypothesis. This helps us to limit the chances of making a false positive conclusion.

Determining Statistical Significance: The significance level determines the threshold for statistical significance. If the p-value is less than or equal to the significance level, the results are considered statistically significant, meaning that the observed effect is unlikely to have occurred by chance alone.

Common Significance Levels

The most commonly used significance levels are 0.05 (5%) and 0.01 (1%).

0.05 Significance Level: This means that there is a 5% chance of rejecting the null hypothesis when it is actually true. In other words, we are willing to accept a 5% risk of making a Type I error.

0.01 Significance Level: This means that there is only a 1% chance of rejecting the null hypothesis when it is actually true. This provides a higher level of confidence in our conclusions but also makes it more difficult to reject the null hypothesis.

Choosing the Significance Level

The choice of significance level depends on the specific research question and the consequences of making a Type I error. In situations where a Type I error could have serious consequences, a lower significance level (e.g., 0.01) may be preferred. In other cases, a higher significance level (e.g., 0.05) may be acceptable.

In Summary

The significance level is an important parameter in hypothesis testing that helps us control the risk of making a Type I error and determine the statistical significance of our results. Choosing the appropriate significance level is essential for drawing valid conclusions from our research.


4. What does a P-value represent in hypothesis testing?
   - Interpretation

Low p-value (typically ≤ 0.05): A low p-value suggests that the observed data is unlikely to have occurred by chance alone if the null hypothesis were true. This provides evidence against the null hypothesis, leading to its rejection in favor of the alternative hypothesis.
High p-value (typically > 0.05): A high p-value suggests that the observed data is likely to have occurred by chance alone if the null hypothesis were true. This means there is not enough evidence to reject the null hypothesis.
Relationship to Significance Level

The p-value is compared to the significance level (alpha) to make a decision about the null hypothesis.

If p-value ≤ alpha: Reject the null hypothesis.
If p-value > alpha: Fail to reject the null hypothesis.
Important Considerations

P-value is not the probability that the null hypothesis is true. It only represents the probability of observing the data if the null hypothesis were true.
A low p-value does not necessarily mean the alternative hypothesis is true. It only suggests that the null hypothesis is unlikely.
The p-value should be interpreted in conjunction with other factors, such as the effect size and the context of the research question.
Example

Imagine testing whether a coin is fair (null hypothesis: the coin is fair). You flip the coin 10 times and get 9 heads. The p-value would be the probability of getting 9 or more heads (or 9 or more tails) if the coin were truly fair. This p-value would be very low, suggesting the coin might not be fair.

In Summary

The p-value is a crucial element in hypothesis testing. It helps assess the strength of evidence against the null hypothesis and guides the decision-making process. However, it's essential to remember its limitations and interpret it carefully in the broader research context.

6. What are Type 1 and Type 2 errors in hypothesis testing?
  - Okay, let's discuss Type 1 and Type 2 errors in hypothesis testing:

Type 1 Error (False Positive)

Definition: A Type I error occurs when you reject the null hypothesis when it is actually true.
Analogy: Think of it like a fire alarm going off when there's no fire. It's a false alarm.
Consequences: You might conclude that there is an effect or relationship when there really isn't one. This can lead to wasted resources or incorrect decisions.
Probability: The probability of making a Type I error is denoted by alpha (α), which is the significance level you set for your hypothesis test.
Example: If you set α = 0.05, there's a 5% chance of making a Type I error.
Type 2 Error (False Negative)

Definition: A Type II error occurs when you fail to reject the null hypothesis when it is actually false.
Analogy: Think of it like a fire alarm not going off when there is a fire. It's a missed detection.
Consequences: You might conclude that there is no effect or relationship when there really is one. This can lead to missed opportunities or incorrect conclusions.
Probability: The probability of making a Type II error is denoted by beta (β).
Example: If you fail to detect a real difference between two groups, you've made a Type II error.
Relationship between Type I and Type II Errors

There's an inherent trade-off between Type I and Type II errors. If you decrease the risk of one type of error, you generally increase the risk of the other.

Lowering α (significance level): Reduces the chance of a Type I error but increases the chance of a Type II error.
Increasing sample size: Can help reduce the risk of both types of errors.
Summary Table

Error Type	Description	Decision	Reality
Type I (False Positive)	Rejecting a true null hypothesis	Reject H0	H0 is true
Type II (False Negative)	Failing to reject a false null hypothesis	Fail to reject H0	H0 is false
In Simple Terms

Type I Error: Saying something is true when it's actually false.
Type II Error: Saying something is false when it's actually true.

8. @ What is the Z-test, and when is it used in hypothesis testing?

  - What is the Z-test?

The Z-test is a statistical test used to determine whether two population means are different when the variances are known and the sample size is large. It is based on the Z-statistic, which follows the standard normal distribution under the null hypothesis.

When is it Used?

The Z-test is used in hypothesis testing when:

The population variance is known. This means that you have prior knowledge about the variability of the data in the population.
The sample size is large (typically n ≥ 30). This is because the central limit theorem ensures that the sampling distribution of the mean will be approximately normal for large sample sizes.
Applications of the Z-test

The Z-test can be used for a variety of hypothesis testing scenarios, including:

Testing the mean of a single population: For example, you might want to test whether the average height of students in a school is significantly different from the national average.
Testing the difference between two population means: For example, you might want to compare the average scores of two groups of students on a test to see if there is a significant difference.
Testing the proportion of a single population: For example, you might want to test whether the proportion of voters who support a particular candidate is significantly different from 50%.
Testing the difference between two population proportions: For example, you might want to compare the proportion of men and women who prefer a certain brand of product.
How to Perform a Z-test

To perform a Z-test, you need to:

State the null and alternative hypotheses.
Set the significance level (alpha).
Calculate the Z-statistic.
Determine the p-value.
Make a decision based on the p-value.
Example

Suppose you want to test whether the average height of students in a school is significantly different from the national average of 5'9". You collect a sample of 100 students and find that their average height is 5'10". You also know that the population standard deviation is 3 inches.

Null Hypothesis: The average height of students in the school is equal to the national average (μ = 5'9").

Alternative Hypothesis: The average height of students in the school is different from the national average (μ ≠ 5'9").

You would then calculate the Z-statistic and determine the p-value. If the p-value is less than your chosen significance level (e.g., 0.05), you would reject the null hypothesis and conclude that the average height of students in the school is significantly different from the national average.

Estimating Population Parameters: Confidence intervals give us a range of values within which we are confident the true population parameter lies. For example, a 95% confidence interval for the population mean might be (10, 12). This means that we are 95% confident that the true population mean falls between 10 and 12.

Assessing Statistical Significance: Confidence intervals can be used to assess the statistical significance of a difference between two groups or a change over time. If the confidence interval for the difference between two means does not include zero, then we can conclude that there is a statistically significant difference between the two groups.

Determining Sample Size: Confidence intervals can be used to determine the sample size needed to achieve a desired level of precision in an estimate. For example, if we want to estimate the population mean with a margin of error of 1, we can use a confidence interval calculation to determine the required sample size.

Example

Suppose we want to estimate the average height of women in a population. We collect a sample of 100 women and find that their average height is 5'4" with a standard deviation of 2 inches. We calculate a 95% confidence interval for the population mean to be (5'3", 5'5").

Interpretation: We are 95% confident that the true average height of women in the population falls between 5'3" and 5'5". This means that if we were to repeat the study many times, 95% of the confidence intervals we calculate would contain the true population mean.

Key Points to Remember

Confidence intervals provide a range of plausible values for the true population parameter.

Confidence intervals are a valuable tool for interpreting statistical results and drawing meaningful conclusions from data.

1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results

```python
import math
import scipy.stats as stats

def z_test_one_sample(sample_mean, pop_mean, pop_std, sample_size, alpha=0.05):
    """
    Performs a one-sample Z-test.

    Args:
        sample_mean: Sample mean.
        pop_mean: Population mean.
        pop_std: Population standard deviation.
        sample_size: Sample size.
        alpha: Significance level (default: 0.05).

    Returns:
        z_statistic: Calculated Z-statistic.
        p_value: P-value of the test.
        decision: Decision based on the p-value ('Reject H0' or 'Fail to reject H0').
    """

    # Calculate the Z-statistic
    z_statistic = (sample_mean - pop_mean) / (pop_std / math.sqrt(sample_size))

    # Calculate the p-value (two-tailed test)
    p_value = 2 * stats.norm.sf(abs(z_statistic))  # sf is the survival function (1 - cdf)

    # Make a decision based on the p-value
    decision = "Reject H0" if p_value <= alpha else "Fail to reject H0"

    return z_statistic, p_value, decision

# Example usage
sample_mean = 110  # Sample mean
pop_mean = 100    # Population mean
pop_std = 15      # Population standard deviation
sample_size = 30   # Sample size

z_statistic, p_value, decision = z_test_one_sample(sample_mean, pop_mean, pop_std, sample_size)

print("Z-statistic:", z_statistic)
print("P-value:", p_value)
print("Decision:", decision)

# Interpretation:
if decision == "Reject H0":
    print("There is sufficient evidence to conclude that the sample mean is significantly different from the population mean.")
else:
    print("There is not enough evidence to conclude that the sample mean is significantly different from the population mean.")
```

4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def two_tailed_z_test(sample_mean, pop_mean, pop_std, sample_size, alpha=0.05):
  """
  Performs a two-tailed Z-test and visualizes the decision region.

  Args:
    sample_mean: The sample mean.
    pop_mean: The population mean.
    pop_std: The population standard deviation.
    sample_size: The sample size.
    alpha: The significance level (default: 0.05).
  """

  # Calculate the Z-statistic
  z_statistic = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))

  # Calculate the critical Z-values
  critical_z = stats.norm.ppf(1 - alpha / 2)  # ppf is the percent point function (inverse of cdf)

  # Calculate the p-value
  p_value = 2 * stats.norm.sf(abs(z_statistic))

  # Make a decision
  decision = "Reject H0" if p_value <= alpha else "Fail to reject H0"

  # Visualize the decision region
  x = np.linspace(-4, 4, 1000)  # Range of Z-values for the plot
  y = stats.norm.pdf(x)        # Probability density function of the standard normal distribution

  plt.figure(figsize=(8, 6))
  plt.plot(x, y, label='Standard Normal Distribution')

  # Shade the rejection regions
  plt.fill_between(x, y, where=(x < -critical_z) | (x > critical_z), color='red', alpha=0.3, label='Rejection Region')

  # Mark the critical Z-values and the Z-statistic
  plt.axvline(-critical_z, color='black', linestyle='--', label='Critical Z')
  plt.axvline(critical_z, color='black', linestyle='--')
  plt.axvline(z_statistic, color='green', label='Z-statistic')

  plt.title('Two-Tailed Z-Test Visualization')
  plt.xlabel('Z-statistic')
  plt.ylabel('Probability Density')
  plt.legend()
  plt.grid(True)
  plt.show()

  # Print results
  print("Z-statistic:", z_statistic)
  print("P-value:", p_value)
  print("Critical Z-value:", critical_z)
  print("Decision:", decision)

# Example usage:
sample_mean = 110  # Sample mean
pop_mean = 100    # Population mean
pop_std = 15      # Population standard deviation
sample_size = 30   # Sample size
alpha = 0.05      # Significance level

two_tailed_z_test(sample_mean, pop_mean, pop_std, sample_size, alpha)
```

5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing@

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def visualize_errors(pop_mean_null, pop_mean_alt, pop_std, sample_size, alpha, beta):
  """
  Calculates and visualizes Type I and Type II errors for a one-sample Z-test.

  Args:
    pop_mean_null: Population mean under the null hypothesis.
    pop_mean_alt: Population mean under the alternative hypothesis.
    pop_std: Population standard deviation.
    sample_size: Sample size.
    alpha: Significance level (Type I error rate).
    beta: Type II error rate.
  """

  # Calculate critical Z-values for alpha
  critical_z_lower = stats.norm.ppf(alpha / 2)
  critical_z_upper = stats.norm.ppf(1 - alpha / 2)

  # Calculate critical sample means for alpha
  critical_sample_mean_lower = pop_mean_null + critical_z_lower * (pop_std / np.sqrt(sample_size))
  critical_sample_mean_upper = pop_mean_null + critical_z_upper * (pop_std / np.sqrt(sample_size))

  # Calculate Z-value for beta
  z_beta = stats.norm.ppf(1 - beta)

  # Calculate critical sample mean for beta
  critical_sample_mean_beta = pop_mean_alt - z_beta * (pop_std / np.sqrt(sample_size))

  # Create plot
  x = np.linspace(pop_mean_null - 4 * (pop_std / np.sqrt(sample_size)),
                  pop_mean_alt + 4 * (pop_std / np.sqrt(sample_size)), 1000)

  # Null hypothesis distribution
  y_null = stats.norm.pdf(x, pop_mean_null, pop_std / np.sqrt(sample_size))
  plt.plot(x, y_null, label='Null Hypothesis Distribution')
  plt.fill_between(x, y_null, where=(x < critical_sample_mean_lower) | (x > critical_sample_mean_upper),
                   color='red', alpha=0.3, label='Type I Error (alpha)')

  # Alternative hypothesis distribution
  y_alt = stats.norm.pdf(x, pop_mean_alt, pop_std / np.sqrt(sample_size))
  plt.plot(x, y_alt, label='Alternative Hypothesis Distribution')
  plt.fill_between(x, y_alt, where=(x < critical_sample_mean_beta),
                   color='blue', alpha=0.3, label='Type II Error (beta)')

  plt.title('Type I and Type II Errors')
  plt.xlabel('Sample Mean')
  plt.ylabel('Probability Density')
  plt.legend()
  plt.grid(True)
  plt.show()

# Example usage:
pop_mean_null = 100    # Population mean under null hypothesis
pop_mean_alt = 105     # Population mean under alternative hypothesis
pop_std = 15           # Population standard deviation
sample_size = 30        # Sample size
alpha = 0.05           # Significance level (Type I error rate)
beta = 0.20            # Type II error rate

visualize_errors(pop_mean_null, pop_mean_alt, pop_std, sample_size, alpha, beta)
```

6. Write a Python program to perform an independent T-test and interpret the results

```python
import numpy as np
import scipy.stats as stats

def independent_t_test(group1_data, group2_data, alpha=0.05):
    """
    Performs an independent samples t-test.

    Args:
        group1_data: Data for the first group (list or numpy array).
        group2_data: Data for the second group (list or numpy array).
        alpha: Significance level (default: 0.05).

    Returns:
        t_statistic: Calculated t-statistic.
        p_value: P-value of the test.
        decision: Decision based on the p-value ('Reject H0' or 'Fail to reject H0').
    """

    # Perform the independent t-test
    t_statistic, p_value = stats.ttest_ind(group1_data, group2_data)

    # Make a decision based on the p-value
    decision = "Reject H0" if p_value <= alpha else "Fail to reject H0"

    return t_statistic, p_value, decision

# Example usage:
group1_data = [10, 12, 11, 13, 10, 9, 12, 11, 10, 11]  # Data for group 1
group2_data = [12, 14, 13, 15, 12, 11, 14, 13, 12, 13]  # Data for group 2

t_statistic, p_value, decision = independent_t_test(group1_data, group2_data)

print("T-statistic:", t_statistic)
print("P-value:", p_value)
print("Decision:", decision)

# Interpretation:
if decision == "Reject H0":
    print("There is sufficient evidence to conclude that there is a significant difference between the means of the two groups.")
else:
    print("There is not enough evidence to conclude that there is a significant difference between the means of the two groups.")
```

```python
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):
    """
    Calculates the confidence interval for a sample mean.

    Args:
        data: The sample data (list or numpy array).
        confidence: The confidence level (default: 0.95).

    Returns:
        lower_bound: The lower bound of the confidence interval.
        upper_bound: The upper bound of the confidence interval.
    """

    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Use ddof=1 for sample standard deviation
    sample_size = len(data)

    # Calculate the critical value (t-distribution for small samples, z-distribution for large samples)
    if sample_size < 30:
        critical_value = stats.t.ppf((1 + confidence) / 2, df=sample_size - 1)  # t-distribution
    else:
        critical_value = stats.norm.ppf((1 + confidence) / 2)  # z-distribution

    # Calculate the margin of error
    margin_of_error = critical_value * (sample_std / np.sqrt(sample_size))

    # Calculate the confidence interval bounds
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example usage:
data = [10, 12, 11, 13, 10, 9, 12, 11, 10, 11]  # Sample data
lower, upper = confidence_interval(data, confidence=0.95)

print("Confidence Interval:", (lower, upper))
```

```python
import numpy as np

def bayesian_inference(prior, likelihood, data):
    """
    Performs Bayesian inference using Bayes' Theorem.

    Args:
        prior: Prior probabilities for each hypothesis (list or numpy array).
        likelihood: Likelihood of observing the data given each hypothesis (list or numpy array).
        data: The observed data.

    Returns:
        posterior: Posterior probabilities for each hypothesis.
    """

    # Calculate the evidence (marginal likelihood)
    evidence = np.sum(likelihood * prior)

    # Calculate posterior probabilities using Bayes' Theorem
    posterior = (likelihood * prior) / evidence

    return posterior

# Example usage:
# Hypotheses: H1 (coin is fair), H2 (coin is biased)
prior = np.array([0.5, 0.5])  # Prior probabilities for H1 and H2
likelihood = np.array([0.5, 0.8])  # Likelihood of observing heads given H1 and H2
data = 1  # Observed data (1 for heads, 0 for tails)

posterior = bayesian_inference(prior, likelihood, data)

print("Posterior Probabilities:", posterior)
```

```python
import numpy as np
import pandas as pd

def calculate_expected_frequencies(observed_data):
    """
    Calculates the expected frequencies for a Chi-square test.

    Args:
        observed_data: Observed frequencies (Pandas DataFrame or 2D numpy array).

    Returns:
        expected_frequencies: Expected frequencies (numpy array).
    """

    # Convert to numpy array if needed
    if isinstance(observed_data, pd.DataFrame):
        observed_data = observed_data.to_numpy()

    # Calculate row and column totals
    row_totals = observed_data.sum(axis=1)
    column_totals = observed_data.sum(axis=0)
    grand_total = observed_data.sum()

    # Calculate expected frequencies
    expected_frequencies = np.zeros_like(observed_data, dtype=float)
    for i in range(observed_data.shape[0]):
        for j in range(observed_data.shape[1]):
            expected_frequencies[i, j] = (row_totals[i] * column_totals[j]) / grand_total

    return expected_frequencies

# Example usage:
# Sample observed data (replace with your actual data)
observed_data = pd.DataFrame({
    'Category1': [10, 20, 30],
    'Category2': [15, 25, 35]
})

# Calculate expected frequencies
expected_frequencies = calculate_expected_frequencies(observed_data)

# Print the expected frequencies
print("Expected Frequencies:\n", expected_frequencies)
```

16. Implement an F-test using Python to compare the variances of two random samples

```python
import numpy as np
import scipy.stats as stats

def f_test_variance(data1, data2):
    """
    Performs an F-test to compare the variances of two random samples.

    Args:
        data1: The first sample data (list or numpy array).
        data2: The second sample data (list or numpy array).

    Returns:
        f_statistic: The F-statistic.
        p_value: The p-value of the test.
    """

    # Calculate variances
    variance1 = np.var(data1, ddof=1)  # Use ddof=1 for sample variance
    variance2 = np.var(data2, ddof=1)

    # Calculate the F-statistic
    f_statistic = variance1 / variance2 if variance1 >= variance2 else variance2 / variance1

    # Calculate the degrees of freedom
    df1 = len(data1) - 1
    df2 = len(data2) - 1

    # Calculate the p-value (two-tailed test)
    p_value = 2 * min(stats.f.cdf(f_statistic, df1, df2), 1 - stats.f.cdf(f_statistic, df1, df2))

    return f_statistic, p_value

# Example usage:
# Sample data
data1 = [10, 12, 11, 13, 10, 9, 12, 11, 10, 11]
data2 = [12, 14, 13, 15, 12, 11, 14, 13, 12, 13]

# Perform the F-test
f_statistic, p_value = f_test_variance(data1, data2)

# Print the results
print("F-statistic:", f_statistic)
print("P-value:", p_value)

# Interpretation
alpha = 0.05  # Significance level
if p_value <= alpha:
    print("There is sufficient evidence to reject the null hypothesis.")
    print("The variances of the two samples are significantly different.")
else:
    print("There is not enough evidence to reject the null hypothesis.")
    print("The variances of the two samples are not significantly different.")
```

```python
import pandas as pd
import statsmodels.stats.api as sms
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def check_anova_assumptions(data, dependent_variable, independent_variable):
    """
    Checks the assumptions for ANOVA: normality, independence, and equal variance.

    Args:
        data: Pandas DataFrame containing the data.
        dependent_variable: Name of the dependent variable.
        independent_variable: Name of the independent variable (groups).
    """

    groups = data[independent_variable].unique()

    print("\n--- Normality Assumption (Shapiro-Wilk Test) ---")
    for group in groups:
        group_data = data[data[independent_variable] == group][dependent_variable]

        # Shapiro-Wilk test for normality (only if there are enough samples)
        if len(group_data) >= 3:
            statistic, p_value = stats.shapiro(group_data)
            print(f"Group {group}: Statistic={statistic:.3f}, p-value={p_value:.3f}")
        else:
            print(f"Group {group}: Not enough data points for Shapiro-Wilk test (<3 samples).")

        # Q-Q Plot for Normality Check
        plt.figure(figsize=(6, 4))
        stats.probplot(group_data, plot=plt)
        plt.title(f"Q-Q Plot for Group {group}")
        plt.xlabel("Theoretical Quantiles")
        plt.ylabel("Sample Quantiles")
        plt.show()

    print("\n--- Independence Assumption ---")
    print("Independence should be ensured through experimental design.")

    print("\n--- Homogeneity of Variance (Levene's Test) ---")
    try:
        # Levene's test for equal variance across groups
        statistic, p_value = stats.levene(*[data[data[independent_variable] == group][dependent_variable] for group in groups])
        print(f"Levene's Test: Statistic={statistic:.3f}, p-value={p_value:.3f}")

        # Box Plot for Visualizing Variance
        plt.figure(figsize=(8, 6))
        sns.boxplot(x=independent_variable, y=dependent_variable, data=data, palette="Set2")
        plt.title("Box Plot of Group Variances")
        plt.show()

    except ValueError as e:
        print(f"Levene's test could not be performed: {e}")

# Example usage:
data = pd.DataFrame({
    'Group': ['A', 'A', 'B', 'B', 'C', 'C'],
    'Value': [10, 12, 15, 17, 20, 22]
})

# Check ANOVA assumptions
check_anova_assumptions(data, 'Value', 'Group')
```

24.  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

```python
import numpy as np
import scipy.stats as stats

def variance_test(data, hypothesized_variance, alpha=0.05):
    """
    Performs a hypothesis test for population variance using a Chi-square distribution.

    Args:
        data: The sample data (list or numpy array).
        hypothesized_variance: The hypothesized population variance.
        alpha: The significance level (default is 0.05).

    Returns:
        chi2_statistic: The Chi-square statistic.
        p_value: The p-value of the test.
        decision: The decision ("Reject H0" or "Fail to reject H0").
    """

    n = len(data)
    sample_variance = np.var(data, ddof=1)  # Use ddof=1 for sample variance

    # Calculate the Chi-square statistic
    chi2_statistic = (n - 1) * sample_variance / hypothesized_variance

    # Calculate the p-value (two-tailed test)
    p_value = 2 * min(stats.chi2.cdf(chi2_statistic, n - 1), 1 - stats.chi2.cdf(chi2_statistic, n - 1))

    # Make a decision
    decision = "Reject H0" if p_value <= alpha else "Fail to reject H0"

    return chi2_statistic, p_value, decision

# Example usage:
data = [10, 12, 11, 13, 10, 9, 12, 11, 10, 11]
hypothesized_variance = 2.5

chi2_statistic, p_value, decision = variance_test(data, hypothesized_variance)

# Print the results
print("Chi-square Statistic:", chi2_statistic)
print("P-value:", p_value)
print("Decision:", decision)

# Interpretation:
# - Null Hypothesis (H0): The population variance is equal to the hypothesized variance.
# - Alternative Hypothesis (H1): The population variance is not equal to the hypothesized variance.

# Based on the decision, you can conclude:
# - If the decision is "Reject H0", there is sufficient evidence to conclude that the population variance is different from the hypothesized variance.
# - If the decision is "Fail to reject H0", there is not enough evidence to conclude that the population variance is different from the hypothesized variance.
```

```python
import numpy as np
import statsmodels.stats.proportion as proportion

def ztest_proportions(data1, data2, alternative='two-sided'):
    """
    Performs a Z-test for comparing proportions between two datasets or groups.

    Args:
        data1: A list or numpy array representing the first dataset (0s and 1s).
        data2: A list or numpy array representing the second dataset (0s and 1s).
        alternative: The alternative hypothesis ('two-sided', 'smaller', or 'larger').
                      Defaults to 'two-sided'.

    Returns:
        z_statistic: The Z-statistic.
        p_value: The p-value of the test.
    """

    n1 = len(data1)
    n2 = len(data2)

    # Calculate the number of successes (1s) in each dataset
    successes1 = sum(data1)
    successes2 = sum(data2)

    # Calculate the proportions
    prop1 = successes1 / n1
    prop2 = successes2 / n2

    # Perform the Z-test
    z_statistic, p_value = proportion.proportions_ztest(
        count=[successes1, successes2],
        nobs=[n1, n2],
        alternative=alternative
    )

    return z_statistic, p_value

# Example usage:
data1 = [1, 0, 1, 1, 0, 0, 1, 0, 1, 1]  # Sample data for group 1
data2 = [0, 1, 0, 0, 1, 1, 0, 1, 0, 0]  # Sample data for group 2

z_statistic, p_value = ztest_proportions(data1, data2)

# Print the results
print("Z-statistic:", z_statistic)
print("P-value:", p_value)

# Interpretation:
# - Null Hypothesis (H0): The proportions are equal between the two groups.
# - Alternative Hypothesis (H1): The proportions are not equal (two-sided),
#                                or one proportion is smaller/larger than the other (one-sided).

# If the p-value is less than your significance level (e.g., 0.05), you reject the null hypothesis
# and conclude that there is a significant difference in proportions between the two groups.
```

26.  Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def f_test_variances(data1, data2, alpha=0.05):
    """
    Performs an F-test for comparing the variances of two datasets.

    Args:
        data1: A list or numpy array representing the first dataset.
        data2: A list or numpy array representing the second dataset.
        alpha: The significance level (default is 0.05).

    Returns:
        f_statistic: The F-statistic.
        p_value: The p-value of the test.
        decision: The decision ("Reject H0" or "Fail to reject H0").
    """

    # Calculate variances
    var1 = np.var(data1, ddof=1)  # Use ddof=1 for sample variance
    var2 = np.var(data2, ddof=1)

    # Calculate the F-statistic
    f_statistic = var1 / var2 if var1 >= var2 else var2 / var1

    # Calculate the p-value (two-tailed test)
    dfn = len(data1) - 1
    dfd = len(data2) - 1
    p_value = 2 * min(stats.f.cdf(f_statistic, dfn, dfd), 1 - stats.f.cdf(f_statistic, dfn, dfd))

    # Make a decision
    decision = "Reject H0" if p_value <= alpha else "Fail to reject H0"

    return f_statistic, p_value, decision

# Example usage
data1 = [10, 12, 11, 13, 10, 9, 12, 11, 10, 11]
data2 = [15, 17, 16, 18, 15, 14, 17, 16, 15, 16]

f_statistic, p_value, decision = f_test_variances(data1, data2)

# Print the results
print("F-statistic:", f_statistic)
print("P-value:", p_value)
print("Decision:", decision)

# Interpretation
print("\nInterpretation:")
print("- Null Hypothesis (H0): The variances of the two datasets are equal.")
print("- Alternative Hypothesis (H1): The variances of the two datasets are not equal.")
print(f"- Decision: {decision}")
if decision == "Reject H0":
    print("  - There is sufficient evidence to conclude that the variances are different.")
else:
    print("  - There is not enough evidence to conclude that the variances are different.")

# Visualization
plt.figure(figsize=(8, 6))
sns.boxplot(data=[data1, data2], notch=True)  # Notch shows the confidence interval around the median
plt.title("Comparison of Variances (Boxplots)")
plt.xticks([0, 1], ['Dataset 1', 'Dataset 2'])
plt.ylabel("Values")
plt.show()
```

27.  Perform a Chi-square test for goodness of fit with simulated data and analyze the results.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# 1️⃣ Generate Simulated Data:
np.random.seed(42)  # Set seed for reproducibility
sample_size = 200  # Number of data points
lambda_param = 4    # Poisson distribution parameter

# Generate Poisson-distributed data
simulated_data = np.random.poisson(lambda_param, size=sample_size)

# 2️⃣ Define Expected Frequencies:
categories = np.arange(0, max(simulated_data) + 1)  # Dynamically set categories

# Calculate expected probabilities using Poisson PMF
expected_probabilities = stats.poisson.pmf(categories, lambda_param)

# Normalize to match sample size
expected_frequencies = expected_probabilities * sample_size

# Ensure minimum expected frequency (avoid categories with < 5 expected count)
mask = expected_frequencies >= 5
filtered_categories = categories[mask]
filtered_expected_frequencies = expected_frequencies[mask]

# 3️⃣ Compute Observed Frequencies:
# Use bincount and ensure alignment with expected categories
observed_frequencies = np.bincount(simulated_data, minlength=len(categories))

# Filter observed frequencies using the same mask
filtered_observed_frequencies = observed_frequencies[mask]

# 🔹 Fix: Rescale expected frequencies to match observed sum
filtered_expected_frequencies *= filtered_observed_frequencies.sum() / filtered_expected_frequencies.sum()

# 4️⃣ Perform Chi-square Goodness-of-Fit Test:
chi2_statistic, p_value = stats.chisquare(filtered_observed_frequencies, filtered_expected_frequencies)

# 5️⃣ Analyze Results:
print("\nChi-square Statistic:", chi2_statistic)
print("P-value:", p_value)

alpha = 0.05  # Significance level
print("\nInterpretation:")
print("- Null Hypothesis (H0): The data follows the Poisson distribution.")
print("- Alternative Hypothesis (H1): The data does not follow the Poisson distribution.")

if p_value <= alpha:
    print(f"- Decision: Reject H0 (p-value ≤ {alpha})")
    print("  - The data does NOT fit the Poisson distribution.")
else:
    print(f"- Decision: Fail to reject H0 (p-value > {alpha})")
    print("  - The data is consistent with the Poisson distribution.")

# 6️⃣ Plot Observed vs. Expected Frequencies:
plt.figure(figsize=(8, 5))
plt.bar(filtered_categories, filtered_observed_frequencies, alpha=0.7, label="Observed", color="blue")
plt.plot(filtered_categories, filtered_expected_frequencies, 'ro-', markersize=8, label="Expected", linewidth=2)
plt.xlabel("Poisson Distributed Categories")
plt.ylabel("Frequency")
plt.legend()
plt.title("Observed vs. Expected Poisson Distribution")
plt.show()
```
